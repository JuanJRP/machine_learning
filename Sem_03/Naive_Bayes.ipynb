# **Naive Bayes Classifier**

```Iris``` dataset description [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris) \
```GaussianNB``` documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html#sklearn.naive_bayes.GaussianNB)

````
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris # Iris dataset
````

````
iris = load_iris() # Load the iris dataset
iris_X = iris.data[:, 0:2] # Features. We'll pick 2
iris_y = iris.target # Labels

# Print shapes
print(iris_X.shape, iris_y.shape)
````

````
_, ax = plt.subplots(figsize=(6,5))
ax.scatter(iris_X[:, 0], iris_X[:, 1], c=iris_y, cmap='Set1', edgecolors='k')
ax.set_xlabel('Sepal length')
ax.set_ylabel('Sepal width')
````

````
# Instance Bayes object
naive_bayes = GaussianNB()

# Training
naive_bayes.fit(iris_X, iris_y)
````

````
## Plot decision boundaries
# Min-Max values for the grid of decision boundaries
offeset = 0.2
x_min, x_max = min(iris_X[:,0]) - offeset, max(iris_X[:,0]) + offeset
y_min, y_max = min(iris_X[:,1]) - offeset , max(iris_X[:,1]) + offeset

# Create grid
x_grid, y_grid = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
print(x_grid.shape, y_grid.shape)
print(x_grid.ravel().shape, y_grid.ravel().shape)
print(np.concatenate([x_grid.ravel().reshape(-1,1), y_grid.ravel().reshape(-1,1)], axis=1).shape)
print(np.c_[x_grid.ravel(), y_grid.ravel()].shape)

# Estimate predictions over the grid
pred_grid = naive_bayes.predict(np.c_[x_grid.ravel(), y_grid.ravel()])
print (pred_grid.shape)

# Reshape predictions
pred_grid = pred_grid.reshape(x_grid.shape)
print (pred_grid.shape)
````

````
# Draw grid
_, ax = plt.subplots (figsize=(6,5))
ax.pcolormesh( x_grid , y_grid , pred_grid , cmap = "Set1")

# Scatter real values
ax.scatter ( iris_X[:,0] , iris_X[:,1] , c = iris_y , edgecolors = 'k', cmap = "Set1")

plt.tight_layout()
ax.set_xticks(())
ax.set_yticks(())
````
