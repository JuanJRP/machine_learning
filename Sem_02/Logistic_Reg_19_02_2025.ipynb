# **In Synthetic**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples = 500 , n_features = 2 , n_redundant = 0, n_informative = 2, n_clusters_per_class = 1, n_classes = 2, random_state = 8)
print(X.shape)
print(y.shape)
_, ax = plt.subplots(figsize=( 5, 4 ))
ax.scatter( X[:,0] , X[:,1] , c = y , edgecolors='k', cmap = 'Set1')
```

```python
# Create an instance
logistic_regression = LogisticRegression()

# Train the model
logistic_regression.fit ( X , y ) #(# Samples, # Features)
```

```python
print(logistic_regression.coef_ , logistic_regression.intercept_)
```

# **Decision boundaries**

```python
# Min-Max values for the grid of decision boundaries
offeset = 0.2
x_min, x_max = min(X[:,0]) - offeset, max(X[:,0]) + offeset
y_min, y_max = min(X[:,1]) - offeset , max(X[:,1]) + offeset

# Create grid
x_grid, y_grid = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
print(x_grid.shape, y_grid.shape)
print(x_grid.ravel().shape, y_grid.ravel().shape)
print(np.concatenate([x_grid.ravel().reshape(-1,1), y_grid.ravel().reshape(-1,1)], axis=1).shape)
print(np.c_[x_grid.ravel(), y_grid.ravel()].shape)

# Estimate predictions over the grid
pred_grid = logistic_regression.predict(np.c_[x_grid.ravel(), y_grid.ravel()])
print (pred_grid.shape)

# Reshape predictions
pred_grid = pred_grid.reshape(x_grid.shape)
print (pred_grid.shape)
```

```python
# Draw grid
_, ax = plt.subplots (figsize=(6,5))
ax.pcolormesh( x_grid , y_grid , pred_grid , cmap = "Set1")

# Scatter real values
ax.scatter ( X[:,0] , X[:,1] , c = y , edgecolors = 'k', cmap = "Set1")

ax.set_xticks(())
ax.set_yticks(())
```
