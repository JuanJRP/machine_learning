## 🔥 Exercise: Identify the Type of Learning 🤖

**🎯 Objective:** Write a Python function that classifies a given problem into Supervised, Unsupervised, or Semi-supervised Learning based on the dataset description.

**📝 Instructions:**

1. Define a function ```identify_learning_type(description: str)```.

2. The function should check for keywords such as:
  - 📌 *Supervised Learning* → If the description contains any of these words:
  "labels", "target", "ground truth", "training examples", "annotation", the function returns "Supervised Learning".
  - 📌 *Unsupervised Learning* → If it contains:
  "clusters", "patterns", "grouping", "hidden structure", "data distribution", the function returns "Unsupervised Learning".
  - 📌 *Semi-supervised Learning* → If it contains:
  "few labels", "unlabeled data", "limited supervision", "pseudo-labels", "partially annotated", the function returns "Semi-supervised Learning".
  
3. If none of these keywords are found, the function returns "Unknown Learning Type".


🚀 Test the function with a different example dataset descriptions.

  ```
  # Example Tests
  print(identify_learning_type("This dataset has input features and corresponding labels."))

  print(identify_learning_type("The goal is to find hidden structure and patterns in the data."))
  ```

````python
def identify_learning_type(description: str):
    if "labels" in description or "target" in description or "ground truth" in description or "training examples" in description or "annotation" in description:
        return "Supervised Learning"

    elif "clusters" in description or "patterns" in description or "grouping" in description or "hidden structure" in description or "data distribution" in description:
        return "Unsupervised Learning"

    elif "few labels" in description or "unlabeled data" in description or "limited supervision" in description or "pseudo-labels" in description or "partially annotated" in description:
        return "Semi-supervised Learning"
````

````python
text = input("Add a description: ")
identify_learning_type(text)
````
