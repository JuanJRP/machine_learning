## In Synthetic data

````python
import numpy as np # Librer√≠a mateamtica
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import make_regression
````

````python
X, y = make_regression(n_samples = 300, n_features = 1, noise = 20, random_state = 1)

_, ax = plt.subplots (figsize=(6,5))
# ax.set_xlim(-3 , -1)
# ax.set_ylim(125, -50)
ax.scatter( X , y , c = 'b', edgecolors="k")
````

````python
svr = SVR(kernel = "linear" )

svr.fit( X , y )
````

````python
# Predecir la nueva muestra. Imprimir la prediccion
prediction = svr.predict( np.array( [ -2 ] ).reshape(-1, 1) )
print (prediction)
````

````python
# Predict x values
predictions = svr.predict( X )

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , color = 'blue', edgecolors="k")
ax.plot ( X , predictions , color = 'red', linestyle = "--")
````

# Non-linear data

````python
import numpy as np # Librer√≠a mateamtica
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import make_regression
````

````python
X, y = make_regression(n_samples = 300, n_features = 1, noise = 20, random_state = 1)
y = (y ** 2) * 0.001
_, ax = plt.subplots (figsize=(6,5))
ax.scatter( X , y , c = 'b', edgecolors="k")

print(X.shape)
print(y.shape)
````

````python
svr = SVR(kernel = "rbf", C = 100)

svr.fit( X , y )
````

````python
# Predict x values
predictions = svr.predict( X )

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , color = 'blue', edgecolors="k")
ax.scatter ( X , predictions , color = 'red', s= 5, linestyle = "--")
print(X.shape)
print(y.shape)
````
